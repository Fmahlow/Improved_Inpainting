```python
import gradio as gr
import transformers
import numpy as np
import torch
from skimage import io
from diffusers import AutoPipelineForText2Image, AutoPipelineForInpainting, StableDiffusionXLPipeline
import torch
from PIL import Image, ImageDraw
import cv2
import pandas as pd
```

```python
pipe_seg = transformers.pipeline("image-segmentation", model="briaai/RMBG-1.4", trust_remote_code=True)
pipe_fundo = AutoPipelineForText2Image.from_pretrained("stabilityai/sdxl-turbo", torch_dtype=torch.float16, variant="fp16").to("cuda")
pipe_inpainting = pipeline = AutoPipelineForInpainting.from_pretrained("runwayml/stable-diffusion-inpainting", torch_dtype=torch.float16, variant="fp16")
pipe_inpainting.enable_model_cpu_offload()
pipe_inpainting.enable_vae_tiling()
```

```python
def generate_background(foreground_image, prompt_fundo):
    return pipe_fundo(prompt=prompt_fundo, num_inference_steps=1, guidance_scale=0.0, height=foreground_image.size[1], width=foreground_image.size[0]).images[0]

def remove_background(img_path, return_mask=False):
    return pipe_seg(img_path, return_mask=return_mask)

def replace_background(foreground_image, background_image, foreground_mask):
    foreground_pixels = foreground_image.load()
    background_pixels = background_image.load()
    mask_pixels = foreground_mask.load()
    for y in range(foreground_mask.size[1]):
        for x in range(foreground_mask.size[0]):
            if mask_pixels[x, y] == 0:
                foreground_pixels[x, y] = background_pixels[x, y]

    return foreground_image

def detect_faces(foreground_image, foreground_mask):
    foreground_image_np = np.array(foreground_image)
    new_mask = np.array(foreground_mask.copy())
    blue_mask = np.zeros((new_mask.shape[0], new_mask.shape[1], 3), dtype=np.uint8)

    
    gray = cv2.cvtColor(foreground_image_np, cv2.COLOR_BGR2GRAY)
    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
    faces = face_cascade.detectMultiScale(gray, minSize=(100, 100))
    
    for (x, y, w, h) in faces:
        face_img = Image.fromarray(foreground_image_np[y:y+h, x:x+w])
        center = (x + w // 2, y + h // 2)
        axes = (int(w * 0.5), int(h * 0.8))
        angle = 0
        color = (0, 0, 0)
        thickness = -1
    cv2.ellipse(new_mask, center, axes, angle, 0, 360, color, thickness)
    cv2.ellipse(blue_mask, center, axes, angle, 0, 360, (0, 0, 255), -1)
    
    face_coordinates = []
    for i in range(blue_mask.shape[0]):
        for j in range(blue_mask.shape[1]):
            if np.all(blue_mask[i, j] == [0, 0, 255]):
                face_coordinates.append([i, j])
    
    return Image.fromarray(new_mask), pd.DataFrame(face_coordinates)

def change_clothes(clothes_prompt, image_with_new_background, mask_with_face_rec):
    return pipe_inpainting(prompt=clothes_prompt, image=image_with_new_background, mask_image=mask_with_face_rec).images[0]

def run_pipeline(img_path, prompt_background, clothes_prompt):
    foreground_image = remove_background(img_path, return_mask=False)
    new_width = foreground_image.size[0] - (foreground_image.size[0] % 8)
    new_height = foreground_image.size[1] - (foreground_image.size[1] % 8)
    foreground_image = foreground_image.crop((0, 0, new_width, new_height))
    foreground_mask = remove_background(img_path, return_mask=True).crop((0, 0, new_width, new_height))
    background_image = generate_background(foreground_image, prompt_background)
    mask_with_face_rec, coords = detect_faces(foreground_image, foreground_mask)
    image_with_new_background = replace_background(foreground_image, background_image,foreground_mask)
    new_background_and_clothes = np.array(change_clothes(clothes_prompt, image_with_new_background.convert('RGB'), mask_with_face_rec.convert('RGB')))
    foreground_image_np = np.array(foreground_image.convert('RGB'))
    new_background_and_clothes[coords[0], coords[1]] = foreground_image_np[coords[0], coords[1]]
    return Image.fromarray(new_background_and_clothes)
```

```python
def change_model(choice):
  model_config = {
      "Midas-Profundidade": (False, True),
      "OWLv2-SAM": (True, True),
      "*": (False, False)
  }
  text_visible, slider_visible = model_config.get(choice, model_config["*"])
  text = gr.Textbox(lines=8, label="Sujeito a ser segmentado", visible=text_visible, interactive=True)
  slider_thr = gr.Slider(0, 100, 0, label="Threshrod", visible=slider_visible, interactive=True)

  return text, slider_thr
```

```python
with gr.Blocks() as demo:
    gr.Markdown("Pipeline para substituiÃ§Ã£o da roupas/ fundo de maneira automÃ¡tica")
    with gr.Row():
        with gr.Column():
            with gr.Row():
                with gr.Column():
                    fundo = gr.Textbox(label="Digite o prompt para o fundo", placeholder="Digite aqui o prompt")
                with gr.Column():
                    roupas = gr.Textbox(label="Digite o prompt para as roupas", placeholder="Digite aqui o prompt")
                models = ["BRIA-RMBG-1.4","Midas-Profundidade","OWLv2-SAM"]
                valor_radio = gr.Radio(models, label='Escolha um modelo:', value="BRIA-RMBG-1.4")
                text = gr.Textbox(lines=8, label="Sujeito a ser segmentado", visible = False, interactive = True)
                slider_thr = gr.Slider(0, 100, 0, label="Threshold", visible = False, interactive = True)
                valor_radio.change(fn=change_model, inputs=valor_radio, outputs=[text, slider_thr])
            imagem = gr.Image(type="filepath")
            btn = gr.Button(value="Trocar Fundo/roupas")
        with gr.Column():
            output_image = gr.Image(interactive=False)
            btn.click(run_pipeline, inputs=[imagem, fundo, roupas], outputs=output_image)

demo.launch(debug=True, share=True)
```
